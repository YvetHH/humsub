```python
subspecies_importance = pd.read_csv("../LODO/subspecies_importance.tsv",sep='\t')
subspecies_importance.Feature = subspecies_importance.Feature.astype(str).str.rjust(10,'0')
```

```python
median_importance =subspecies_importance.groupby(['Feature']).median().drop('random_state',axis=1)
```

```python
median_importance = median_importance.reset_index()
```

```python
median_importance = median_importance.sort_values(by='Value',ascending=False)
```

```python
colors = ["darkred", "red", "lightcoral"]  # dark red for low, light red for high

# Create a custom colormap
custom_cmap = sns.color_palette("dark:salmon_r", as_cmap=True)

```

```python


# Adjust the 'Feature' column and create 'cluster95'
subspecies_importance['Feature'] = subspecies_importance['Feature'].astype(str).str.rjust(10, '0')
subspecies_importance['cluster95'] = subspecies_importance['Feature'].str[:4]

# Compute median value for each feature in each test study across random states
median_values = subspecies_importance.groupby(['test_study', 'Feature']).median()['Value'].reset_index()

# Rank the features within each test_study
median_values['Rank'] = median_values.groupby('test_study')['Value'].rank(method='min', ascending=False)

# Merge the 'cluster95' information into the median_values DataFrame
median_values_with_cluster = pd.merge(median_values, subspecies_importance[['Feature', 'cluster95']], on='Feature', how='left')

# Drop duplicates that might have been introduced by the merge
median_values_with_cluster = median_values_with_cluster.drop_duplicates()

# Correcting the extraction of top 20 features per test study
top_20_features_per_study = median_values_with_cluster.groupby('test_study').apply(
    lambda x: x.nsmallest(30, 'Rank', 'all')['Feature']
).reset_index()

# Extract and flatten the list of top 20 features
top_20_features_flattened = top_20_features_per_study['Feature'].tolist()

# Correcting the extraction of top 50 'cluster95' per test study
top_50_cluster95_per_study = median_values_with_cluster.groupby('test_study').apply(
    lambda x: x.nsmallest(50, 'Rank', 'all')['Feature']
).reset_index()
top_50_cluster95_per_study['cluster95'] = top_50_cluster95_per_study['Feature'].str[:4]
# Extract and flatten the list of top 50 'cluster95'
top_50_cluster95_flattened = top_50_cluster95_per_study['cluster95'].tolist()

# Filter to get the features whose 'cluster95' is not among the top 50 of the corresponding test study
features_of_interest = []
for feature in top_20_features_flattened:
    # Find the corresponding test study for this feature
    test_studies = top_20_features_per_study[top_20_features_per_study['Feature'] == feature]['test_study']
    if len(test_studies) > 1:
        for test_study in test_studies:
            # Get the cluster95 value of this feature
            cluster95 = feature[:4]
            # Check if this cluster95 is not in the top 50 of the test study
            if top_50_cluster95_per_study[top_50_cluster95_per_study['test_study'] == test_study]['cluster95'].value_counts().loc[cluster95] == 1:
                features_of_interest.append(feature)
        else:
            test_study = test_studies.iloc[0]
            cluster95 = subspecies_importance[subspecies_importance['Feature'] == feature]['cluster95'].iloc[0]
            # Check if this cluster95 is not in the top 50 of the test study
            if top_50_cluster95_per_study[top_50_cluster95_per_study['test_study'] == test_study]['cluster95'].value_counts().loc[cluster95] == 1:
                features_of_interest.append(feature)
# features_of_interest
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Compute median value for each feature in each test study across random states
median_values = subspecies_importance.groupby(['test_study', 'Feature']).median()['Value'].reset_index()

# Rank the features within each test_study
median_values['Rank'] = median_values.groupby('test_study')['Value'].rank(method='min', ascending=False)

# Identify top 20 features for each test_study
top_20_features_per_study = median_values.groupby('test_study').apply(lambda x: x.nsmallest(30, 'Rank')['Feature']).reset_index()

# Flatten the list of top 20 features
top_20_features_flattened = top_20_features_per_study['Feature'].unique()

# Subset the original rankings to include only the top 20 features
subset_rankings = median_values[median_values['Feature'].isin(top_20_features_flattened)]

# Pivot the table for the clustermap
subset_pivot_table_top_20 = subset_rankings.pivot(index='Feature', columns='test_study', values='Rank')

# Fill NaN values with a large number to indicate a low rank
filled_pivot_table_top_20 = subset_pivot_table_top_20.fillna(100)

# Create the clustermap with the filled pivot table for top 20 features
filled_pivot_table_top_20 = filled_pivot_table_top_20.loc[(filled_pivot_table_top_20 <= 30).sum(1) > 1]
g=sns.clustermap(filled_pivot_table_top_20, cmap=custom_cmap, annot=True,vmin=1,vmax=60,fmt='g',figsize=(6, 12),
                dendrogram_ratio=(0.3, 0.05))
for label in g.ax_heatmap.get_yticklabels():
    if label.get_text() in red_labels_translated:
        label.set_color('red')
        
        
g.ax_heatmap.set_xlabel('')
g.ax_heatmap.set_ylabel('')
plt.setp(g.ax_heatmap.get_xticklabels(), fontsize=14,rotation=90) 
plt.setp(g.ax_heatmap.get_yticklabels(), fontsize=18,fontstyle='italic')
#g.savefig("subspecies_clustermap.svg",bbox_inches='tight')
plt.show()
```
