```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import roc_auc_score, f1_score, accuracy_score, recall_score, confusion_matrix

from lightgbm import LGBMClassifier

from shapmat.abundance_filter import ab_filter

from statannotations.Annotator import Annotator
```

```python
### SUBSPECIES-LEVEL DATA ###

sample_metadata = pd.read_csv('../data/CRC_metadata.tsv',sep='\t',index_col=0)
sample_metadata = sample_metadata.query("study_condition != 'adenoma'")


relab_subsp = pd.read_csv("../data/relab_subsp.tsv",sep='\t',index_col=0)
relab_subsp.index = relab_subsp.index.astype(str).str.rjust(10,'0')
relab_subsp = relab_subsp.T
relab_subsp = relab_subsp.loc[relab_subsp.index.isin(sample_metadata.index)]
sample_metadata = sample_metadata.loc[relab_subsp.index]
```

```python
### SPECIES-LEVEL DATA ###

relab_sp = pd.read_csv("../data/relab_sp.tsv",sep='\t',index_col=0).T

relab_sp = relab_sp.loc[relab_sp.index.isin(sample_metadata.index)]
sample_metadata = sample_metadata.loc[sample_metadata.index.isin(relab_sp.index)]

name_mapping = dict(zip(relab_sp.columns, range(len(relab_sp.columns))))
relab_sp.columns = relab_sp.columns.map(name_mapping)
```

```python
relab_sp = relab_sp.loc[sample_metadata.index]
relab_subsp = relab_subsp.loc[sample_metadata.index]
```

```python
from sklearn.model_selection import KFold, StratifiedKFold
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
import numpy as np

def perform_k_fold_cv(data, labels, params, k=5, random_state=None,data_type='subspecies'):
    """
    Perform k-fold cross-validation using LightGBM.

    Parameters:
    - data: DataFrame with samples as rows and features as columns.
    - labels: Series with target labels for each sample.
    - metadata: Series with index as sample_ids matching the data's rows.
    - params: Dictionary with parameters for the LGBMClassifier.
    - k: Number of folds for cross-validation.
    - random_state: Random state for reproducibility.

    Returns:
    - mean_accuracy: Mean accuracy across all folds.
    - fold_accuracies: List of accuracies for each fold.
    """
    kf = StratifiedKFold(n_splits=k, shuffle=True, random_state=random_state)
    feature_importance = pd.DataFrame()
    output_df = pd.DataFrame()
    for train_index, test_index in kf.split(data, labels):
        # Splitting the data and labels according to the current fold
        X_train, X_test = data.iloc[train_index], data.iloc[test_index]
        y_train, y_test = labels.iloc[train_index], labels.iloc[test_index]
        
        y_train = y_train.map({"control":0,"CRC":1})
        y_test = y_test.map({"control":0,"CRC":1})
        
        X_train_filtered = ab_filter(X_train,abundance_threshold=0,prevalence_threshold=0.8)
        X_test_filtered = X_test[X_train_filtered.columns]
        
        # Initializing and training the LGBM model
        lgbm = LGBMClassifier(**params, n_estimators=1000, random_state=random_state)
        lgbm.fit(X_train_filtered, y_train)

        # Predicting and evaluating the model
        y_pred_proba = lgbm.predict_proba(X_test_filtered)[:, 1]
        roc_auc = roc_auc_score(y_test, y_pred_proba)
        
        y_pred = lgbm.predict(X_test_filtered)

        accuracy = accuracy_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        confusion_matrix_score = confusion_matrix(y_test, y_pred)
        ppv = confusion_matrix_score[1,1] / confusion_matrix_score[1,:].sum()
        npv = confusion_matrix_score[0,0] / confusion_matrix_score[0,:].sum()
        
        output_df = pd.concat([output_df, pd.DataFrame({"model":"LBGM", 'data_type':data_type,"roc_auc":roc_auc,
                                                       "accuracy":accuracy, "recall":recall, "f1_score":f1, 'ppv':ppv, 'npv':npv},
                                                       index=[data_type])], axis=0)
        
        single_model_importance = pd.DataFrame({'Value':lgbm.feature_importances_,'Feature':X_train_filtered.columns, 'data_type':data_type,'random_state':random_state})
        feature_importance = pd.concat([feature_importance, single_model_importance], axis=0)
    
    return output_df, feature_importance
```

```python
params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': 'binary_logloss',
    'num_leaves': 11,
    'learning_rate': 0.01,
    'verbose': 0
}
```

```python
random_states =  [5570, 7954, 2080, 6549, 3410, 8773, 9730, 120, 4821, 2959, 4274, 5855, 6497, 9919, 3151, 8414, 3298, 1026, 3744, 9186, 4736, 4478, 7290, 7141, 2322, 701, 6874, 9946, 1616, 1164]
subsp_performance = pd.DataFrame()
subsp_importance = pd.DataFrame()
for random_state in random_states:
    subsp_performance_tmp, subsp_importance_tmp = perform_k_fold_cv(relab_subsp,sample_metadata['study_condition'],params,10,random_state,data_type='subspecies')
    subsp_performance = pd.concat([subsp_performance,subsp_performance_tmp],axis=0)
    subsp_importance = pd.concat([subsp_importance,subsp_importance_tmp],axis=0)
```
