```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import os, glob
```

```python
def concat_benchmark(input_dir):
    output_df = pd.DataFrame()
    for file in glob.glob(f"{input_dir}/*.txt"):
        sample = file.split("/")[-1].strip(".txt")
        df = pd.read_csv(file, sep='\t')
        df['sample'] = sample
        output_df = pd.concat([output_df,df])
    output_df = output_df.set_index("sample")
    return output_df
```

```python
metaphlan = concat_benchmark("metaphlan/")

sourmash_sketch = concat_benchmark("sourmash/sketching/")
sourmash_gather = concat_benchmark("sourmash/gather/")

sourmash_total = sourmash_gather + sourmash_sketch

sourmash_total['tool'] = 'sourmash'
metaphlan['tool'] = 'metaphlan'
```

```python
total_performance = pd.concat([sourmash_total,metaphlan])
```

```python
total_performance = total_performance.reset_index()
```

```python
total_performance
```

```python
ORDER = ['metaphlan','sourmash']
```

```python
sns.set_context('paper')
fig,ax = plt.subplots(figsize=(6,8))
sns.boxplot(data = total_performance, x='tool',y='s',ax=ax,color='lightgray', linewidth=1.5,order=ORDER)
sns.swarmplot(data = total_performance, x='tool',y='s',ax=ax,s=10,hue='sample',order=ORDER)
sns.lineplot(x="tool", y="s", hue='sample', data=total_performance,estimator=None, legend=False)
#plt.title('L2 distance between reference and test abundances from HumGut MAGs.')
plt.xlabel("")
plt.ylabel("Runtime [s]",fontsize=28)
plt.xticks(fontsize=20,rotation=30)
plt.yticks(fontsize=20)
ax.set_xticklabels(['MetaPhlAn4\n [8 cores]','Sourmash with \nsubspecies catalog'])

ax.get_legend().remove()
#plt.savefig("runtime.svg")
```

```python
sns.set_context('paper')
fig,ax = plt.subplots(figsize=(6,8))
sns.boxplot(data = total_performance, x='tool',y='max_uss',ax=ax,color='lightgray', linewidth=1.5,order=ORDER)
sns.swarmplot(data = total_performance, x='tool',y='max_uss',ax=ax,s=10,hue='sample',order=ORDER)
sns.lineplot(x="tool", y="max_uss", hue='sample', data=total_performance,estimator=None, legend=False)
#plt.title('L2 distance between reference and test abundances from HumGut MAGs.')
plt.xlabel("")
plt.ylabel("Total memory [MB]",fontsize=28)
plt.xticks(fontsize=20, rotation=30)
plt.yticks(fontsize=20)
ax.set_xticklabels(['MetaPhlAn4\n [8 cores]','Sourmash with \nsubspecies catalog'])
ax.get_legend().remove()
#plt.savefig("memory.svg")
```
