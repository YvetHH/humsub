```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

from scipy.stats import pearsonr, spearmanr
```

```python
plt.rcParams['svg.fonttype'] = 'none'
```

```python
def scale_to_0_1(series):
    min_val = series.min()
    max_val = series.max()
    scaled_series = (series - min_val) / (max_val - min_val)
    return scaled_series
```

```python
Tax = pd.read_csv("../meta_analysis/effect_size/tax.tsv",sep='\t',index_col=0)
Tax.cluster95 = Tax.cluster95.astype(str).str.rjust(4,'0')
Tax.index = Tax.index.astype(str).str.rjust(10,'0')
Tax.subspecies = Tax.subspecies.astype(str).str.rjust(6,'0')
Tax.Label = Tax.Label.str.lstrip("s__")

species_tax = Tax[['cluster95','species']]
species_tax.columns = ['cluster95','Label']
species_tax = species_tax.set_index('cluster95')

Tax = pd.concat([Tax,species_tax]).Label

Tax = Tax.loc[~Tax.duplicated()]
Tax = Tax.str.replace(" ", "\n")
```

```python
# Load your data
# Replace 'file_path' with the path to your data file
data = pd.read_csv('countries_comparison.tsv', sep='\t')
data.subspecies = data.subspecies.astype(str).str.rjust(10,'0')

# Original Enrichment Score Calculation
data['weighted_meandiff'] = data['meandiff'].abs() / data['p-adj']
original_enrichment_scores = data.groupby('subspecies')['weighted_meandiff'].mean()
original_normalized_scores = (original_enrichment_scores - original_enrichment_scores.min()) / (original_enrichment_scores.max() - original_enrichment_scores.min())

# Confidence Interval Adjusted Score Calculation
data['interval_width'] = data['upper'] - data['lower']
data['interval_adjusted_score'] = data['weighted_meandiff'] / data['interval_width']
conf_interval_scores = data.groupby('subspecies')['interval_adjusted_score'].mean()
conf_interval_normalized_scores = (conf_interval_scores - conf_interval_scores.min()) / (conf_interval_scores.max() - conf_interval_scores.min())

# Exponential Weighting Score Calculation
data['exp_weight'] = np.exp(-data['p-adj'])
data['exp_interval_adjusted_score'] = data['meandiff'].abs() * data['exp_weight'] / data['interval_width']
exp_weight_scores = data.groupby('subspecies')['exp_interval_adjusted_score'].mean()
exp_weight_normalized_scores = (exp_weight_scores - exp_weight_scores.min()) / (exp_weight_scores.max() - exp_weight_scores.min())

# Composite Score Calculation
data['significance_score'] = 1 / (data['p-adj'] + 0.0001)
prevalence_variability_score = data.groupby('subspecies').size().rename('variability_score')
data = data.join(prevalence_variability_score, on='subspecies', how='left')
data['composite_score'] = (data['meandiff'].abs() + data['significance_score'] + data['variability_score']) / 3
composite_scores = data.groupby('subspecies')['composite_score'].mean()

# Combine all scores
all_scores = pd.DataFrame({
    'Original': original_normalized_scores,
    'Confidence Interval': conf_interval_normalized_scores,
    'Exponential Weight': exp_weight_normalized_scores,
    'Composite': composite_scores
})
all_scores = all_scores['Composite']
# Selecting top and bottom 25% for each score type
top_25 = all_scores.quantile(0.8)
bottom_25 = all_scores.quantile(0.2)

high_specificity = all_scores[(all_scores >= top_25)]
low_specificity = all_scores[(all_scores <= bottom_25)]

# Calculating median differences and p-values for high and low specificity groups
median_diff_high = data[data['subspecies'].isin(high_specificity.index)].groupby('subspecies')['meandiff'].median()
median_diff_low = data[data['subspecies'].isin(low_specificity.index)].groupby('subspecies')['meandiff'].median()
p_value_high = data[data['subspecies'].isin(high_specificity.index)].groupby('subspecies')['p-adj'].median()
p_value_low = data[data['subspecies'].isin(low_specificity.index)].groupby('subspecies')['p-adj'].median()
```

```python
# Plotting
plt.figure(figsize=(12, 5))
sns.set_context('paper')
# Plot for median differences
plt.subplot(1, 2, 1)
sns.histplot(median_diff_high, color="blue", label="High Specificity", kde=False, binwidth=0.01)
sns.histplot(median_diff_low, color="red", label="Low Specificity", kde=False, binwidth=0.01)
#plt.title("Distribution of Median Difference")
plt.xlabel("Median Difference", fontsize=16)
plt.ylabel("Frequency",fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
#plt.ylim(0,2000)
plt.legend(fontsize=14)
#plt.savefig("median_difference.svg")
```

```python
# Plot for p-values
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
sns.histplot(p_value_high, color="blue", label="High Specificity", kde=False, binwidth=0.01)
sns.histplot(p_value_low, color="red", label="Low Specificity", kde=False, binwidth=0.01)
plt.xlabel("Median p-value", fontsize=16)
plt.ylabel("Frequency",fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
#plt.ylim(0,2000)
#plt.legend()

plt.tight_layout()
#plt.savefig("p_value.svg")
plt.show()
```

```python
score_vs_p_adj = pd.concat([data[['p-adj','subspecies']].groupby('subspecies').median()['p-adj'],all_scores],axis=1)
```
